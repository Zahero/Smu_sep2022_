* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

```python
tot_schools=len(school_data)
tot_students=len(student_data)
tot_budget=school_data.budget.sum()
avg_reading=student_data.reading_score.mean()
avg_math=student_data.math_score.mean()


```

```python
mask=student_data.math_score>=70
num_pass_math=len(student_data.loc[mask])
#sum(mask)
math_perc=100*(num_pass_math/tot_students)
math_perc
```

```python
mask=student_data.reading_score>=70
num_pass_reading=len(student_data.loc[mask])
#sum(mask)
reading_perc=100*(num_pass_reading/tot_students)
reading_perc
```

```python
mask=(student_data.reading_score>=70) & (student_data.math_score>=70)
num_pass_both=len(student_data.loc[mask])
#sum(mask)
both_perc=100*(num_pass_both/tot_students)
both_perc
```

```python
output = {
    "Total Schools": tot_schools,
    "Total Students": tot_students,
    "Total Budget": tot_budget,
    'Average Math Score': avg_math,
    "Average Reading Score": avg_reading,
    "% Passing Math": math_perc,
    "% Passing Reading": reading_perc,
    "% Overall Passing": both_perc
}
summ1 = pd.DataFrame([output])
summ1
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
df = school_data_complete.copy()
```

```python
df.head()
```

```python
s1 = df.groupby(["school_name", "type"]).student_id.count()
s1
```

```python
s2 = school_data.groupby(["school_name", "type"]).budget.sum()
s2
```

```python
s3 = s2 / s1
s3
```

```python
s4 = df.groupby(["school_name", "type"]).math_score.mean()
s4
```

```python
s5 = df.groupby(["school_name", "type"]).reading_score.mean()
s5
```

```python
passed_math = df.loc[df.math_score >= 70].groupby(["school_name", "type"]).student_id.count()
s6 = 100*(passed_math / s1)

s6
```

```python
passed_reading = df.loc[df.reading_score >= 70].groupby(["school_name", "type"]).student_id.count()
s7 = 100*(passed_reading / s1)

s7
```

```python
passed_both = df.loc[(df.reading_score >= 70) & (df.math_score >= 70)].groupby(["school_name", "type"]).student_id.count()
s8 =100* (passed_both / s1)

s8
```

```python
summ2 = pd.DataFrame()
summ2["Total Students"] = s1
summ2["Total Budget"] = s2
summ2["Per Student Budget"] = s3
summ2["Average Math"] = s4
summ2["Average Reading"] = s5
summ2["% Passing Math"] = s6
summ2["% Passing Reading"] = s7
summ2["% Passing Both"] = s8

summ2
```

```python
df["pass_math"] = df["math_score"] >= 70
df["pass_reading"] = df["reading_score"] >= 70
df["pass_both"] = (df["math_score"] >= 70) & (df["reading_score"] >= 70)

df.head()
```

```python
summ22 = df.groupby("school_name").agg({"type": "first", "student_id": "count", "budget": "first", "math_score": "mean", "reading_score": "mean", "pass_math": "sum", "pass_reading":"sum", "pass_both": "sum"})
summ22
```

```python
summ22["Per Student"] = summ22.budget / summ22.student_id 
summ22["% Pass math"] = summ22.pass_math / summ22.student_id 
summ22["% Pass reading"] = summ22.pass_reading / summ22.student_id 
summ22["% Pass both"] = summ22.pass_both / summ22.student_id 
summ22
```
